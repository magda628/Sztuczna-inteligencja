```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

print(tf.__version__)
```

```python
training_data = np.load('/content/sample_data/TF_dataset.npz')
input_size = 2
output_size = 1
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(output_size)
])
model.compile(optimizer='sgd', loss='mean_squared_error')
model.fit(training_data['inputs'], training_data['targets'], epochs=100, verbose=1)
```

```python
model.layers[0].get_weights()
```

```python
plt.plot(np.squeeze(model.predict_on_batch(training_data['inputs'])),
         np.squeeze(training_data['targets']), color='orange')
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```

```python
# Przykład z Adam + Huber Loss
model_exp1 = tf.keras.Sequential([
    tf.keras.layers.Dense(output_size)
])

model_exp1.compile(optimizer='adam', loss='huber')

# Poprawione odwołania do danych treningowych
model_exp1.fit(training_data['inputs'], training_data['targets'], epochs=100, verbose=1)

# Poprawione odwołanie przy predykcji
predictions_exp1 = model_exp1.predict_on_batch(training_data['inputs'])

plt.plot(np.squeeze(predictions_exp1), np.squeeze(training_data['targets']), color='blue')
plt.xlabel('outputs')
plt.ylabel('targets')
plt.title('Adam + Huber Loss')
plt.show()
```

```python
model_nonlinear = tf.keras.Sequential([
    tf.keras.layers.Dense(5, activation='sigmoid'),
    tf.keras.layers.Dense(5, activation='sigmoid'),
    tf.keras.layers.Dense(output_size)
])

model_nonlinear.compile(optimizer='sgd', loss='mean_squared_error')

# POPRAWKA: training_data['klucz'], a nie ['klucz']
model_nonlinear.fit(training_data['inputs'], training_data['targets'], epochs=100, verbose=1)

predictions_nl = model_nonlinear.predict_on_batch(training_data['inputs'])

plt.plot(np.squeeze(predictions_nl), np.squeeze(training_data['targets']), color='orange')
plt.xlabel('outputs')
plt.ylabel('targets')
plt.title('Model nieliniowy (sigmoid)')
plt.show()
```

Model liniowy (SGD + MSE) Sieć z jedną warstwą i bez funkcji aktywacji idealnie nauczyła się wzoru. Wagi i biasy po treningu były bardzo bliskie oryginalnym wartościom (2, -3, 5).
Wykres predykcji vs targety to prosta linia — model działa poprawnie.

Różne optymalizatory i funkcje straty Adam zbiega szybciej niż SGD — strata spada wyraźnie już w pierwszych epokach. Huber Loss jest mniej czuła na wartości odstające niż MSE, ale przy czystych danych różnica jest niewielka. Warto eksperymentować, bo dobór tych funkcji zależy od charakteru danych.

Model nieliniowy z sigmoid Sigmoid "ściska" wartości do przedziału (0, 1), przez co sieć nie może dobrze dopasować się do danych o dużym zakresie wartości. Wykres przestaje być prostą — pojawia się rozrzut i wygięcie. Funkcje aktywacji mają sens przy problemach nieliniowych, nie przy prostej regresji liniowej.

Wniosek ogólny Sieć neuronowa potrafi samodzielnie odkryć zależności w danych bez znajomości wzoru. Kluczowe jest jednak dopasowanie architektury i funkcji do typu problemu — zły dobór pogarsza wyniki nawet przy dobrych danych.
